```python
#2. Add Active Cases column in DataFrame

data = df
data['Active'] = data['Confirmed'] - data['Deaths'] - data['Recovered']
data
```

```python
#3. latest number of confirmed, deaths, recovered and active cases of Novel Coronavirus (COVID-19) Country wise

data_country_only = data.groupby(['Country/Region'])[['Confirmed','Deaths','Recovered','Active']].sum()
data_country_only
```

```python
#4. latest number of confirmed, deaths and recovered people of Novel Coronavirus (COVID-19) cases 
#   Country/Region - Province/State wise

data_country_province = data.groupby(['Country/Region','Province/State'])[['Confirmed','Deaths','Recovered']].sum()
data_country_province
```

```python
#5. Get the Chinese province wise cases of confirmed, deaths and recovered cases of Novel Coronavirus (COVID-19)

data_country_province = data[data['Country/Region'] == 'China'].groupby(['Province/State'])[['Confirmed','Deaths','Recovered']].sum() 
data_country_province
```

```python
#6. get the top 10 countries data (Last Update, Country/Region, Confirmed, Deaths, Recovered) 
#   of Novel Coronavirus (COVID-19)

top_10 = data[["Last Update","Country/Region", "Confirmed", "Deaths", "Recovered"]]
top_10.loc[:10]
```

```python
#7. To list countries with all cases of Novel Coronavirus (COVID-19) recover
#   removed all zero

recovered_country = data[["Country/Region","Recovered"]]
recovered_country.loc[(recovered_country!=0).all(axis=1)]
```
